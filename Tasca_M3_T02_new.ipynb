## Troba quantes dades faltants hi ha per columna.

```python
DatosFaltantes = my_df_sampled.isna().sum()
print(DatosFaltantes)
```

### Luego de utilizar el metodo isna() y sumarlos con sum() podemos ver que tenemos datos faltantes, en las columnas , ArrTime , TailNum, ActualElapsedTime, CRSElapsedTime , AirTime, ArrDelay ,  TaxiIn y TaxiOut  ya que estos datos no pueden ser abordados desde una media o un tiempo estimado ademas del data set ser lo suficientemente grande como para poder obtener buenas metricas procederemos a eliminar las filas que contengan registros faltantes.

```python
my_df_cleaned = my_df_sampled.dropna(subset=['ArrTime' , 'TailNum', 'ActualElapsedTime', 'CRSElapsedTime' , 'AirTime', 'ArrDelay' ,  'TaxiIn', 'TaxiOut'])
my_df_cleaned.head(21)
```

```python
check_datos_faltantes = my_df_cleaned.isna().sum()
check_datos_faltantes
```

### De esta manera comprobamos que no hayan datos faltantes ya que sea han eliminado los registros faltantes con el metodo .dropna() 

### A continuacion comprabaremos si hay valores duplicados dentro de nuestro DataSet 

```python
df_cleaned = my_df_cleaned.duplicated()
sum_duplicates = df_cleaned.sum()
print("Cantidad de filas duplicadas:", sum_duplicates)
```

## Crea columnes noves (velocitat mitjana del vol)


```python
my_df_cleaned['VelocidadMediaMph'] = my_df_cleaned['Distance'] / (my_df_cleaned['AirTime'] / 60)
my_df_cleaned.head()
```

### Se ha creado una nueva columna en nuestro DataSet en donde el resultado de la division entre la Distancia (expresada en millas) y el Tiempo de Vuelo expresado en minutos dividido 60, no da como resultado la velocidad media en millas por hora.

## (si ha arribat tard o no) 

```python
my_df_cleaned['Arrivo_Tardio'] = my_df_cleaned['ArrDelay'] >= 15
my_df_cleaned.head()
```

### Se ha creado una columna llamda Arrivo Tardio a partir del calculo de la columna que nos trae el DataSet  (ArrDelay)  en donde se informa de la llega tarde o no de los vuelos, en este caso si el tiempo es mayor a 15 minutos se considera tarde.

## Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
airline_total_delays = my_df_cleaned.groupby('UniqueCarrier')['ArrDelay'].sum()
airline_total_delays = airline_total_delays.sort_values(ascending=False)
airline_total_delays.head(10)
```

### Teniendo como columna UniqueCarrier (nombres de las aerolineas) y ArrDelay (con la cantidad de minutos de retraso acumulados) podemos agrupar las aerolineas con el metodo groupby() y sumar lo minutos con el metodo .sum() para luego ordenar los valores de forma descendiente y asi poder generar un nuevo cuadro con los datos obtenidos. 

## Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
VuelosMasLargos = my_df_cleaned.nlargest(200000, 'Distance')[['FlightNum', 'Distance']]

mas_largos_no_duplicados = VuelosMasLargos.drop_duplicates()

mas_largos_no_duplicados.head(10)
```

### Estos fueron los vuelos mas largos sin repetir el numero de vuelo.

```python
VuelosMasRetrasados = my_df_cleaned.nlargest(10, 'ArrDelay')[['FlightNum', 'Distance', 'ArrDelay']]
mas_retrasados_no_duplicados = VuelosMasRetrasados.drop_duplicates()
mas_retrasados_no_duplicados.head(10)
```

### Estos fueron los 10 vuelos mas retrasados.
